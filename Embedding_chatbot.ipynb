```python
import numpy as np
import time
import json
import os
from sentence_transformers import SentenceTransformer

def create_embedding(text, retry_count=3, delay=1):
    model = SentenceTransformer('all-MiniLM-L6-v2')
    embedding = model.encode(text, show_progress_bar=True)

    return embedding.tolist()

    # for attempt in range(retry_count):
    #     try:
    #         result = genai.embed_content(
    #             model=embedding_model,
    #             content=text,
    #             task_type="SEMANTIC_SIMILARITY"
    #         )
    #         return result["embedding"]
    #     except Exception as e:
    #         if "TooManyRequests" in str(e) and attempt < retry_count - 1:
    #             # Tăng thời gian chờ theo cấp số nhân (exponential backoff)
    #             wait_time = delay * (2 ** attempt)
    #             print(f"Rate limit exceeded. Waiting for {wait_time} seconds...")
    #             time.sleep(wait_time)
    #         else:
    #             raise e

def save_embeddings_to_json(embeddings_dict, filename="intent_embeddings.json"):
    """Lưu embeddings vào file JSON"""
    # Tạo thư mục nếu chưa tồn tại
    os.makedirs(os.path.dirname(filename) if os.path.dirname(filename) else '.', exist_ok=True)

    # Chuyển đổi numpy arrays thành lists để có thể serialize
    serializable_dict = {}
    for key, value in embeddings_dict.items():
        if isinstance(value, np.ndarray):
            serializable_dict[key] = value.tolist()
        elif isinstance(value, list) and all(isinstance(item, np.ndarray) for item in value):
            serializable_dict[key] = [emb.tolist() for emb in value]
        else:
            serializable_dict[key] = value

    with open(filename, 'w') as f:
        json.dump(serializable_dict, f)

    print(f"Embeddings saved to {filename}")

def load_embeddings_from_json(filename="/content/intent_embeddings_data/all_embedded_data.json"):
    """Đọc embeddings từ file JSON"""
    if not os.path.exists(filename):
        return {}

    with open(filename, 'r') as f:
        loaded_dict = json.load(f)

    # Chuyển đổi lists thành numpy arrays
    embeddings_dict = {}
    for key, value in loaded_dict.items():
        if key.endswith("_avg"):
            embeddings_dict[key] = np.array(value)
        else:
            embeddings_dict[key] = [np.array(emb) for emb in value]

    return embeddings_dict

# Tên file để lưu embeddings
embeddings_file = "intent_embeddings.json"

# Kiểm tra xem đã có file embeddings chưa
intent_embeddings = load_embeddings_from_json(embeddings_file)

# Tạo embeddings cho tất cả các câu mẫu với độ trễ và lưu sau mỗi intent
for intent, examples in intents_data.items():
    # Kiểm tra xem intent này đã được xử lý chưa
    if intent in intent_embeddings and intent + "_avg" in intent_embeddings:
        print(f"Intent '{intent}' already processed. Skipping...")
        continue

    print(f"Processing intent: {intent}")
    intent_embeddings[intent] = []

    for i, example in enumerate(examples):
        print(f"  Creating embedding for: {example}")
        try:
            embedding = create_embedding(example)
            intent_embeddings[intent].append(embedding)

            # Lưu ngay sau mỗi embedding để không mất dữ liệu nếu gặp lỗi
            if (i + 1) % 3 == 0 or i == len(examples) - 1:  # Lưu sau mỗi 3 ví dụ hoặc ví dụ cuối cùng
                save_embeddings_to_json(intent_embeddings, embeddings_file)
                print(f"  Saved progress after example {i+1}/{len(examples)}")

            time.sleep(1)  # Thêm độ trễ 1 giây giữa các yêu cầu

        except Exception as e:
            print(f"Error processing example '{example}': {e}")
            # Vẫn lưu những gì đã xử lý được
            if intent_embeddings[intent]:
                save_embeddings_to_json(intent_embeddings, embeddings_file)
            raise e

    # Tính embedding trung bình cho mỗi intent
    if intent_embeddings[intent]:  # Đảm bảo có ít nhất một embedding
        intent_embeddings[intent + "_avg"] = np.mean(intent_embeddings[intent], axis=0)

        # Lưu sau khi tính embedding trung bình
        save_embeddings_to_json(intent_embeddings, embeddings_file)
        print(f"Completed and saved intent: {intent}")

    time.sleep(2)  # Thêm độ trễ lớn hơn giữa các intent

print("All intents processed and saved successfully!")

# Kiểm tra kết quả
for intent in intents_data.keys():
    if intent in intent_embeddings:
        print(f"Intent '{intent}': {len(intent_embeddings[intent])} examples")
        if intent + "_avg" in intent_embeddings:
            print(f"  Average embedding shape: {intent_embeddings[intent + '_avg'].shape}")
```

```python
from typing import Tuple
from sklearn.metrics.pairwise import cosine_similarity

import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from typing import List, Tuple, Dict, Any

def classify_intent_knn(query: str, k: int = 3) -> Tuple[str, float]:
    # Tạo embedding cho câu truy vấn
    query_embedding = create_embedding(query)

    # Chuẩn bị danh sách để lưu trữ tất cả các độ tương đồng
    all_similarities = []

    # Tính độ tương đồng với tất cả các câu mẫu
    for item in intent_embeddings:  # Iterate through the list
        intent = item["tag"]
        example_embedding = np.array(item["embedding"])  # Convert to NumPy array

        similarity = cosine_similarity([query_embedding], [example_embedding])[0][0]
        all_similarities.append({
            "intent": intent,
            "similarity": similarity
        })

    # Sắp xếp theo độ tương đồng giảm dần
    all_similarities.sort(key=lambda x: x["similarity"], reverse=True)

    # Lấy k láng giềng gần nhất
    top_k = all_similarities[:k]

    # Đếm số lần xuất hiện của mỗi intent trong top k
    intent_counts = {}
    for item in top_k:
        intent = item["intent"]
        if intent in intent_counts:
            intent_counts[intent] += 1
        else:
            intent_counts[intent] = 1

    # Tìm intent phổ biến nhất
    if not intent_counts:
        return "unknown", 0.0  # Handle the case where no intents are found

    best_intent = max(intent_counts, key=intent_counts.get)

    # Tính độ tin cậy dựa trên số lần xuất hiện và độ tương đồng trung bình
    confidence = intent_counts[best_intent] / k
    avg_similarity = sum(item["similarity"] for item in top_k if item["intent"] == best_intent) / intent_counts[best_intent]

    # Kết hợp hai yếu tố để có độ tin cậy tổng thể
    overall_confidence = (confidence + avg_similarity) / 2

    return best_intent, overall_confidence

result = classify_intent_knn(query='Sản phẩm X này làm bằng vật liệu gì')

print(result)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def classify_intent(query, threshold=0.75):
    """
    Phân loại intent của câu truy vấn dựa trên embeddings

    Args:
        query: Câu truy vấn của người dùng
        threshold: Ngưỡng độ tương đồng tối thiểu để xác định intent

    Returns:
        intent: Intent được phân loại
        confidence: Độ tin cậy của phân loại
    """
    # Tạo embedding cho câu truy vấn
    query_embedding = create_embedding(query)

    # So sánh với embedding trung bình của mỗi intent
    max_similarity = -1
    best_intent = "unknown"

    for intent in intents_data.keys():
        # Sử dụng embedding trung bình của intent
        avg_embedding = intent_embeddings[intent + "_avg"]

        # Tính độ tương đồng cosine
        similarity = cosine_similarity([query_embedding], [avg_embedding])[0][0]

        if similarity > max_similarity:
            max_similarity = similarity
            best_intent = intent

    # Kiểm tra ngưỡng
    if max_similarity < threshold:
        return "unknown", max_similarity

    return best_intent, max_similarity

```
